### Import the libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
```

### Import the dataset

```python
df = pd.read_excel('images_analyzed.xlsx')
df.head()
```

### Plot the data

```python
sns.lmplot(x='Time', y='Images_Analyzed', data=df, hue='Age')
```

```python
x = df['Time']
y = df['Images_Analyzed']
```

```python
plt.plot(x, y, color="r", alpha=0.5)
plt.xlabel('Time')
plt.ylabel("Images_Analyzed")
plt.title("Inspecting ")
plt.show()
```

```python
sns.lmplot(x='Coffee', y='Images_Analyzed', data=df, hue='Age', order = 2)
```

### Fit the model

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Time', 'Coffee', 'Age']], df.Images_Analyzed)
```

```python
print(reg.coef_)    #coefficients of x1,x2,x3
```

### Predicting

```python
print(reg.predict([[13, 2, 23]]))
```
